# Clustering Crypto

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Fetching Cryptocurrency Data

```python
url = "https://min-api.cryptocompare.com/data/all/coinlist"

```

```python
# Creating a DataFrame
crypto_df = pd.read_csv("crypto_data.csv", index_col=0)
```

```python
crypto_df.head()
```

### Data Preprocessing

```python
# Traspose the data

```

```python
# Remove unnecessary columns
```

```python
# Keep only cryptocurrencies that are on trading
crypto_df = crypto_df[crypto_df['IsTrading']]
crypto_df.head()
```

```python
# Keep only cryptocurrencies with a working algorithm
```

```python
# Remove the "IsTrading" column
crypto_df.drop(columns=['IsTrading'], inplace=True)
crypto_df.head()
```

```python
# Removing rows with at least 1 null value
crypto_df = crypto_df.dropna()
```

```python
# Removing rows with cryptocurrencies without coins mined
crypto_df = crypto_df[crypto_df['TotalCoinsMined']!=0]
```

```python
# Dropping rows where there are 'N/A' text values
crypto_df = crypto_df[(crypto_df['CoinName']!= "N/A") & (crypto_df['Algorithm']!="N/A") & (crypto_df['ProofType']!="N/A")]
```

```python
# Fetching the cryptocurrencies names prior to drop them from crypto_df
coins_name = pd.DataFrame(crypto_df.CoinName)
```

```python
# Removig the cryptocurrency name since it's not going to be used on the clustering algorithm
crypto_df.drop(columns=['CoinName'], inplace=True)
```

```python
# Creating dummies variables for text features

X = pd.get_dummies(crypto_df,columns=['Algorithm', 'ProofType'])
```

```python
# Standardizing data
crypto_scaled = StandardScaler().fit_transform(X)
```

### Reducing Dimensions Using PCA

```python
# Using PCA to reduce dimension to 3 principal components
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
# Creating a DataFrame with the principal components data
pcs_df = pd.DataFrame(
    data=crypto_pca, columns=["PC 1", "PC 2", "PC 3"]
)
pcs_df.head()
```

### Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range ok k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

Running K-Means with `k=<you best value for k here>`

```python
# Initialize the K-Means model
km = KMeans(n_clusters=6, random_state=0)

# Fit the model
km.fit(pcs_df)
# Predict clusters
predictions = km.predict(pcs_df)
# Creating a new DataFrame including predicted clusters and cryptocurrencies features
pcs_df['class'] = km.labels_
pcs_df.index = crypto_df.index
```

```python
crypto_df.head()
clustered_df = pd.concat([crypto_df, pcs_df, coins_name], axis=1, )
clustered_df.head()
```

### Visualizing Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(clustered_df, x="PC 1", y="PC 2", z="PC 3", color="class", hover_name="CoinName", hover_data=["Algorithm"])
fig.show()
```

#### Table of Tradable Cryptocurrencies

```python
# Table with tradable cryptos
clustered_df.hvplot.table(columns=['CoinName', "Algorithm", "ProofType", "TotalCoinSupply", "TotalCoinsMined", "class"])
```
